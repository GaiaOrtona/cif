### D8-D16. Employed graduate (MA) students after 3 years from graduation (2014-2022)


The dataset has been taken from AlmaLaurea and contains data about graduate students after 3 years from their graduation. The dataset takes in consideration only graduate master degree students. It was possible only to download the CSV of the entire enquiry year by year, so a profound process of cleaning and deletion of unnecessary parts of the dataset was undetaken. Due to the strange structure of the original dataset (only one column was present), it was necessary to extract useful data and then recreate a new dataset from scratch. This is the "intermediate dataset" that contains data about the percentage of employed graduates from 2014 to 2022. 

The dataset related to the year 2022 is the only one which contains values that are not decoded correctly. In this case the string "ï¿½" present throughout the column "Scienze della Formazione primaria (corso pre-riforma D.M. n. 249/2010)" means that there is no data available related to that year, therefore this value has been replaced with the string "none".

Furthermore, even though the filtering inside AlmaLaurea website was done considering both the year in which surveys were conducted and the years from graduation, no data about the year in which data was collected was present inside the original datasets. For this reason the additional column "year" has been manually added to each dataset.

```python
import pandas as pd

# Read the CSV file
pd.set_option('display.max_rows', None)
almalaurea2022 = pd.read_csv("C:/Users/crosi/Documents/GitHub/cif/datasets/occupazione_laurea/dati_occupazione_2022.csv", encoding="ISO-8859-1", sep=";")

almalaurea2022.head(5)

```

```python
# Extract rows from index 15 to 40
selected_rows = almalaurea2022.iloc[25:26]
selected_rows1 = almalaurea2022.iloc[29:30]
n = []
z = []
print(selected_rows)
print(selected_rows1)
selected_list = selected_rows.values.tolist()[0]
new = selected_list[0].split(')', 1)[0] + ')'
selected_lis= selected_rows1.values.tolist()[0]
# Print or use selected_rows as needed
columns_to_delete = ['1. Popolazione analizzata']
for column in columns_to_delete:
    del selected_rows[column]
for i in selected_rows.index[0]:
        n.append(i)
nz = n[1:]
nz.append(new)
nz = [item if 'ï¿½' not in item else 'none' for item in nz]

for  m in selected_rows1.index[0]:
    z.append(m)
    
z.append(selected_lis[0])
z = [item if 'ï¿½' not in item else 'none' for item in z]


# Create an empty DataFrame with specified columns
df2022 = pd.DataFrame(columns=nz)
df2022.loc[len(df2022)] = z[1:]
new_column_value = '2022'

# Add a new column with the specified value
df2022['year'] = new_column_value
# Now, df is an empty DataFrame with the specified column names
df2022

```
